## Import Library

```python
import pandas as pd
from keras.models import Sequential
from keras.layers import *
from keras.utils import plot_model
```

## Load Final Training

```python
Data_Learn = pd.read_csv('Final_Training.csv')
Data_Learn.head(2)
```

```python
X = Data_Learn.drop('Rate',axis=1).values
Y = Data_Learn[['Rate']].values
```

```python
Data_Uji = pd.read_csv('Final_Test.csv')
Data_Uji.head(2)
```

```python
X_Uji = Data_Uji.drop('Rate',axis=1).values
Y_Uji = Data_Uji[['Rate']].values
```

# Make Neural Network Model

<img src='example_network.svg'>

```python
model = Sequential()
model.add(Dense(50, input_dim=4, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(1, activation='linear'))
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.fit(
    X,
    Y,
    epochs=800,
    shuffle=True,
    verbose=2
)

Test_Error_Rate = model.evaluate(
    X_Uji,
    Y_Uji,
    verbose=0
)
```
