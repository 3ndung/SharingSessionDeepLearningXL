# Import Library

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

## Load Data_Set

```python
Data_set = pd.read_csv('MMS_Training_Data.csv',sep=';')
Data_set.head(10)
```

## Filtering 

```python
Data_set = Data_set.drop(['MMBAS','CGI','ID','B_Number','Bearer_type','Unit_Type','A_Number','Date','Start_Time','Balance','Scen_Name'],axis=1)
Data_set.head(10)
```

## Make One Hot Encoding

```python
Feat_Data = pd.get_dummies(Data_set, columns=['TYPE'])
Feat_Data
```

```python
Data_Training = Feat_Data.head(28)
Data_Test = Feat_Data.tail(14)
```

## Normalisation

```python
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
Data_Trainig_Scale = scaler.fit_transform(Data_Training)
Data_Test_Scale = scaler.transform(Data_Test)
```

```python
print "Berikut adalah Nilai dari Rate dikali {:.10f} & dibagi {:.6f}".format(scaler.scale_[2], scaler.min_[2])
```

```python
Data_Training_Scale_df = pd.DataFrame(Data_Trainig_Scale, columns=Data_Training.columns.values)
Data_Test_Scale_df = pd.DataFrame(Data_Test_Scale, columns=Data_Test.columns.values)
```

## Saving ke Dalam format csv baru

```python
Data_Training_Scale_df.to_csv('Final_Training.csv', index=False)
Data_Test_Scale_df.to_csv('Final_Test.csv', index=False)
Data_Training_Scale_df.head(1).drop('Rate', axis=1).to_csv('prediksi.csv', index=False)
```
